```python
def correlation_coefficient(patch_edges,template_edges):
    patch_mean = np.mean(patch_edges)
    template_mean = np.mean(template_edges)
    
    patch_var = (patch_edges-patch_mean)
    template_var = (template_edges-template_mean)
    
    correlation = np.sum(utils.elementwise_mul(patch_var,template_var))/np.sqrt(np.sum(np.square(patch_var))*np.sum(np.square(template_var)))
    
    return correlation
```

```python
def convolve2d(img, kernel):
    """Convolves a given image and a given kernel.

    Steps:
        (1) flips the either the img or the kernel.
        (2) pads the img or the flipped img.
            this step handles pixels along the border of the img,
            and makes sure that the output img is of the same size as the input image.
        (3) applies the flipped kernel to the image or the kernel to the flipped image,
            using nested for loop.

    Args:
        img: nested list (int), image.
        kernel: nested list (int), kernel.

    Returns:
        img_conv: nested list (int), image.
    """
    # TODO: implement this function.
    #raise NotImplementedError

    #img = img.tolist()
    img_conv = img.copy()
    flipped_kernel = utils.flip2d(kernel)
    padded_img = utils.zero_pad(img,1,1)

    #print(len(img))
    #print(len(img[0]))

    for i in range(len(img)):
        for j in range(len(img[0])):
            if i+len(kernel) > len(padded_img) or j+len(kernel[0]) > len(padded_img[0]):
                #print(f'{i} and {j} skipped')
                continue
            patch = utils.crop(padded_img,i,i+3,j,j+3)
            filtered_patch = utils.elementwise_mul(patch,kernel)

            pixel_value=0
            #print(len(filtered_patch),len(filtered_patch[0]))
            for k in range(len(filtered_patch)):
                for l in range(len(filtered_patch[0])):
                    pixel_value += filtered_patch[k][l]

            img_conv[i][j] = pixel_value
    img_conv = np.asarray(img_conv)
    return img_conv
```

```python
def save_results(coordinates, template, template_name, rs_directory):
    results = {}
    results["coordinates"] = sorted(coordinates, key=lambda x: x[0])
    results["templat_size"] = (len(template), len(template[0]))
    with open(os.path.join(rs_directory, template_name), "w") as file:
        json.dump(results, file)
```
